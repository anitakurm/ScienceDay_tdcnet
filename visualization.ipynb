```python
import pandas as pd
```

```python
df = pd.read_csv('data/transformed_data_for_viewing.csv')


# df["Occupation"] = df["Occupation"].astype("category", categories=["Data Scientist","Student","Teacher","Nurse","Fire Fighter"], ordered=True)
# df["Occupation"] = df["Occupation"].astype('category')
df["Occupation"] = pd.Categorical(df["Occupation"],["Data Scientist","Student","Teacher","Nurse","Firefighter"])
df["Family Type"] = pd.Categorical(df["Family Type"],["Family", "Couple", "Singles", "Seniors"])
df["Family Interest"] = pd.Categorical(df["Family Interest"], ["Gaming", "Musical", "Crafty", "Sporty", "Travelling", "Outdoor"])
df["House Type"] = pd.Categorical(df["House Type"],["Metal", "Glas", "Concrete", "Tiles", "Wood"])
```

```python
df = df.sort_values("Occupation")
```

```python
import seaborn as sns
pp = sns.pairplot(data=df,
                  y_vars=['Mobile Traffic'],
                  x_vars=['Occupation', 'Family Type', 'Family Interest', 'House Type'])
for ax in pp.axes.flatten():
    
    ax.tick_params(rotation = 90)

pp = sns.pairplot(data=df,
                  y_vars=['Mobile Traffic'],
                  x_vars=['Number Of Computers', 'Number Of Tvs','Number Of Pets', 'Number Of Residents', 'Number Of Phones','Average Age'])


pp = sns.pairplot(data=df,
                  y_vars=['Mobile Traffic'],
                  x_vars=['Customer Happiness', 'Time Spend On YouTube [min]','Time Spend On TikTok [min]','Time Spend On Instagram [min]','Time Spend On Spotify [min]','Size of home [m2]','Distance To Nearest Tower [m]'])

```
